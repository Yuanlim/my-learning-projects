```python
import sys
sys.path.append('../Ds_Module')

# import
from LinkList import LinkList, Node

type T = int;
```

```python
linkList = LinkList[T]()
for i in range(10):
    linkList.AddLast(i)
print(linkList.Count)
print(linkList.Forward())
print(linkList.Backward())
```

```python
# Should be reversal
linkList = LinkList[T]()
for i in range(10):
    linkList.AddFirst(i)
print(linkList.Count)
print(linkList.Forward())
print(linkList.Backward())
```

```python
# Error insert after or before not found node
targetNode:Node[T] = linkList.Find(10)
if (linkList.Find(10) is not None):
  linkList.InsertAfter(linkList.Find(10), 7)

# Brute
linkList.InsertBefore(linkList.Find(10), 6)
```

```python
# None safe test
linkList.InsertAfter(linkList.Find(5), 7) # [9, 8, 7, 6, 5, 7, 4, 3, 2, 1, 0]
linkList.InsertBefore(linkList.Find(8), 10) # [9, 10, 8, 7, 6, 5, 7, 4, 3, 2, 1, 0]
print(f"LinkList: {linkList.Forward()} Count: {linkList.Count}")
```

```python
linkList.RemoveFirst()
linkList.RemoveLast()
print(f"LinkList: {linkList.Forward()} Count: {linkList.Count}")
```

```python
targetNode = linkList.Find(7)
if(targetNode is not None):
  linkList.Remove(targetNode)
print(f"LinkList: {linkList.Forward()} Count: {linkList.Count}")
```
